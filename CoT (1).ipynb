```python
from unsloth import FastLanguageModel, is_bfloat16_supported
import torch
max_seq_length = 1024 # Can increase for longer reasoning traces
lora_rank = 64 # Larger rank = smarter, but slower

model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "Qwen/Qwen2.5-3B-Instruct",
    max_seq_length = max_seq_length,
    load_in_4bit = True, # False for LoRA 16bit
    fast_inference = True, # Enable vLLM fast inference
    max_lora_rank = lora_rank,
    gpu_memory_utilization = 0.5, # Reduce if out of memory
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = lora_rank, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = [
        "q_proj", "k_proj", "v_proj", "o_proj",
        "gate_proj", "up_proj", "down_proj",
    ], # Remove QKVO if out of memory
    lora_alpha = lora_rank,
    use_gradient_checkpointing = "unsloth", # Enable long context finetuning
    random_state = 3407,
)
```

```python
# 3Ô∏è‚É£  Build the prompt exactly as before
SYSTEM_PROMPT = """
Respond in the following format:
<reasoning>
...
</reasoning>
<answer>
...
</answer>
"""
prompt = tokenizer.apply_chat_template(
    [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user",   "content": "How many r's are in strawberry?"},
    ],
    tokenize=False,
    add_generation_prompt=True,
)
```

```python
text = tokenizer.apply_chat_template([
    {"role" : "system", "content" : SYSTEM_PROMPT},
    {"role" : "user", "content" : "How many r's are in strawberry?"},
], tokenize = False, add_generation_prompt = True)

from vllm import SamplingParams
sampling_params = SamplingParams(
    temperature = 0.8,
    top_p = 0.95,
    max_tokens = 1024,
)
output = model.fast_generate(
    text,
    sampling_params = sampling_params,
    lora_request = model.load_lora("grpo_saved_lora"),
)[0].outputs[0].text

output
```

```python
# One-example GSM8K sanity check
%%time
import time, re, string
from datasets import load_dataset
from vllm import SamplingParams

# helpers
def first_int(text):
    m = re.search(r"-?\d+", text)
    return int(m.group()) if m else None

_tbl = str.maketrans("", "", string.punctuation)
def normalise(text):
    return text.lower().translate(_tbl).strip()

def is_correct(gold, pred):
    gnum, pnum = first_int(gold), first_int(pred)
    if gnum is not None and gnum == pnum:
        return True
    return normalise(gold) in normalise(pred)

# attach LoRA once
lora_handle = model.load_lora("grpo_saved_lora")

# fetch one test example
test_ds  = load_dataset("openai/gsm8k", "main", split="test")
ex       = test_ds[1]                       # choose any index
question = ex["question"]
gold_ans = ex["answer"].split("####")[1].strip()

prompt = tokenizer.apply_chat_template(
    [{"role":"system","content":SYSTEM_PROMPT},
     {"role":"user",  "content":question}],
    tokenize=False,
    add_generation_prompt=True,
)

# deterministic generation
sampling = SamplingParams(
    temperature = 0.0,      # 0 ‚Üí greedy / deterministic
    top_p       = 1.0,
    max_tokens  = 200,
)

t0 = time.perf_counter()
completion = model.fast_generate(
    prompt,
    sampling_params = sampling,
    lora_request    = lora_handle,
)[0].outputs[0].text
latency = time.perf_counter() - t0

# extract <answer> block
m = re.search(r"<answer>\s*(.*?)\s*</answer>", completion, re.S)
pred_ans = m.group(1).strip() if m else completion.strip()

# results
print("====== QUESTION ======")
print(question, "\n")
print("====== MODEL OUTPUT ======")
print(completion, "\n")
print("====== GOLD ANSWER ======")
print(gold_ans)
print("====== PREDICTED ANSWER ======")
print(pred_ans)
print(f"Correct? {is_correct(gold_ans, pred_ans)}")
print(f"Latency: {latency:.2f} s")

```

```python
# GSM8K full-test evaluation
%%time
import time, re, string, math
from datasets import load_dataset
from vllm import SamplingParams
from tqdm.auto import tqdm

# helpers
def first_int(text):
    m = re.search(r"-?\d+", text)
    return int(m.group()) if m else None

_tbl = str.maketrans("", "", string.punctuation)
def normalise(text):
    return text.lower().translate(_tbl).strip()

def is_correct(gold, pred):
    gnum, pnum = first_int(gold), first_int(pred)
    if gnum is not None and gnum == pnum:
        return True
    return normalise(gold) in normalise(pred)

# attach LoRA once
lora_handle = model.load_lora("grpo_saved_lora")

# deterministic generation config
sampling = SamplingParams(
    temperature = 0.0,   # greedy / deterministic
    top_p       = 1.0,
    max_tokens  = 200,
)

# Dataset
test_ds = load_dataset("openai/gsm8k", "main", split="test")

# Evaluation loop
correct = 0
latencies = []

for ex in tqdm(test_ds, desc="Evaluating"):
    question = ex["question"]
    gold_ans = ex["answer"].split("####")[1].strip()

    prompt = tokenizer.apply_chat_template(
        [{"role":"system","content":SYSTEM_PROMPT},
         {"role":"user",  "content":question}],
        tokenize=False,
        add_generation_prompt=True,
    )

    tic = time.perf_counter()
    completion = model.fast_generate(
        prompt,
        sampling_params = sampling,
        lora_request    = lora_handle,
    )[0].outputs[0].text
    latencies.append(time.perf_counter() - tic)

    m = re.search(r"<answer>\s*(.*?)\s*</answer>", completion, re.S)
    pred_ans = m.group(1).strip() if m else completion.strip()

    correct += is_correct(gold_ans, pred_ans)

# Summary
N      = len(test_ds)
acc    = correct / N
mean_t = sum(latencies) / N
p95_t  = sorted(latencies)[math.floor(0.95*N)-1]

print("\n‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ Summary ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ")
print(f"Examples evaluated : {N}")
print(f"Exact-match acc    : {acc:.2%}")
print(f"Mean latency (s)   : {mean_t:.2f}")
print(f"95th-pct latency   : {p95_t:.2f}")
```
